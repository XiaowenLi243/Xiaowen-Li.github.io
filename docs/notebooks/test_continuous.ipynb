Analyse results of jobs run using [test_continuous.sl](../slurm/test_continuous.sl).

```python
import pandas as pd
import seaborn as sb
```

```python
cols = ["jobid", "cores", "chunk_size", "time", "memory (GB)"]
jobs = [
    ("47679336", 4, 5, "8min57s", 1.55),
    ("47680561", 4, 20, "6min46s", 1.02),
    ("47681422", 8, 20, "2min55s", 1.67),
    ("47681428", 16, 20, "2min16s", 3.15),
]
dset = pd.DataFrame.from_records(jobs, columns=cols)
dset["time (s)"] = pd.to_timedelta(dset["time"]).dt.total_seconds()
```

```python
dset
```

```python
sb.set()
```

```python
ax = sb.lineplot(
    data=dset, x="cores", y="time (s)", style="chunk_size", markers=True, ms=10
)
ax.set_ylim(bottom=0)
```

```python
ax = sb.lineplot(
    data=dset, x="cores", y="memory (GB)", style="chunk_size", markers=True, ms=10
)
ax.set_ylim(bottom=0)
```
